````python
# A utility function to find all references to an object

def get_references(obj):
    return [ key for key, value in globals().items() if value is obj]


a = "Hello, world!"
b = a
c = "Test string"
d = e = f = c

print(get_references(a))  # Should show 'a' and 'b'
print(get_references(c))  # Should show 'c', 'd', 'e', f
````

````python
class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age

user1 = User("Alice", 30)
user2 = User("Chandrashekar Babu", 49)
print(sys.getsizeof(user1))
print(sys.getsizeof(user2))  # Returns only the shallow size of the object, not including the size of the attributes
````

````python
CREATE_USER_SQL = """
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name VARCHAR(32) UNIQUE NOT NULL,
    role VARCHAR(32)
)"""

INSERT_USER_SQL = """
INSERT INTO users (name, role) VALUES (?, ?)
"""

SELECT_USER_SQL = """
SELECT id, name, role FROM users WHERE role = ?
"""

````

````python
cursor.execute(CREATE_USER_SQL)
````

````python
names = ["Alice", "Bob", "Charlie", "David", "Eve"]
roles = ["admin", "user", "editor", "viewer", "contributor"]
for name, role in zip(names, roles):
    cursor.execute(INSERT_USER_SQL, (name, role))
````

````python
names = ["John", "Jane", "Doe", "Smith", "Emily"]
roles = ["admin", "user", "editor", "viewer", "contributor"]

data = list(zip(names, roles))  # list() used for readability alone.
cursor.executemany(INSERT_USER_SQL, data)
````
