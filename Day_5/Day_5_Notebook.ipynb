````python
def extract_failed_logins(logfile):
    with open(logfile) as ins:
        for line in ins:
            if "Failed password" in line:
                rec = line.split()
                yield {
                    "timestamp": " ".join(rec[:3]),
                    "username": rec[10],
                    "ip_address": rec[12]
                }       

            elif "Invalid user" in line:
                rec = line.split()
                yield {
                    "timestamp": " ".join(rec[:3]),
                    "username": rec[7],
                    "ip_address": rec[9]
                }

import csv
with open("failed_logins.csv", "w") as outs:
    writer = csv.DictWriter(outs, fieldnames=["timestamp", "username", "ip_address"])
    writer.writeheader()
    for record in extract_failed_logins("sshd_minimal.log"):
        writer.writerow(record)
````

````python
class FailedLoginExtractor:
    def __init__(self, logfile):
        self.logfile = logfile
        self.fieldnames = ["timestamp", "username", "ip_address"]

    def extract(self):
        with open(self.logfile) as ins:
            for line in ins:
                if "Failed password" in line:
                    rec = line.split()
                    yield {
                        "timestamp": " ".join(rec[:3]),
                        "username": rec[10],
                        "ip_address": rec[12]
                    }       

                elif "Invalid user" in line:
                    rec = line.split()
                    yield {
                        "timestamp": " ".join(rec[:3]),
                        "username": rec[7],
                        "ip_address": rec[9]
                    }

import csv
failed_logins = FailedLoginExtractor("sshd_minimal.log")

with open("failed_logins.csv", "w") as outs:
    writer = csv.DictWriter(outs, fieldnames=failed_logins.fieldnames)
    writer.writeheader()
    for record in failed_logins.extract():
        writer.writerow(record)
````

#### Weekend mini-project exercise

Implement a chat server and chat client that works fully in command-line.

- The chat server should support multiple clients to connect simultaneously.
- Each client can set a username after connecting and join a chat room.
- When any message is sent by the client, they must broadcast (relayed) to all other
  clients in that room.

In order to support these features, the server must implement the following commands:
``` 
  /list     # list all rooms available
  /join room-name # joins a room
  /leave    # leaves a room
  /name user-name # sets the username
  /users    # list all users in a chat room 
  /quit     # Exit the server
  /help     # list all the above commands with one-line help
```
